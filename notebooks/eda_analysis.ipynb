# Exploratory Data Analysis (EDA)
## Intelligent Reading Comprehension and Quiz Generation System

This notebook performs an exploratory analysis of the reading comprehension dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from collections import Counter

%matplotlib inline
sns.set(style="whitegrid")
```

### 1. Load Data

```python
data_path = '../data/processed/'
train_df = pd.read_csv(os.path.join(data_path, 'train_clean.csv'))

print("Training Data Loaded Successfully!")
```

### 2. Complete Information of Dataset
Checking data types, non-null counts, and memory usage.

```python
print("--- Dataset Info ---")
train_df.info()

print("\n--- Missing Values Before Dropping ---")
print(train_df.isnull().sum())
```

### 3. Drop Null Variables
Removing rows that contain any missing values.

```python
train_df.dropna(inplace=True)
print(f"Missing values after dropping: {train_df.isnull().sum().sum()}")
print(f"New Dataset Shape: {train_df.shape}")
```

### 4. Count and Print Answer Options (A, B, C, D)
Analyzing the distribution of the target variable.

```python
answer_counts = train_df['answer'].value_counts().sort_index()
print("--- Answer Counts ---")
for label, count in answer_counts.items():
    print(f"Option {label}: {count} occurrences")
```

### 5. Plotting the Distributions

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=answer_counts.index, y=answer_counts.values, palette='magma')

plt.title('Distribution of Answer Options (A, B, C, D)', fontsize=15)
plt.xlabel('Answer Option', fontsize=12)
plt.ylabel('Total Count', fontsize=12)

# Adding text labels on bars
for i, val in enumerate(answer_counts.values):
    plt.text(i, val + (max(answer_counts.values)*0.01), str(val), ha='center', fontsize=11)

plt.show()
```

### 6. Text Length Analysis

```python
train_df['article_len'] = train_df['clean_article'].apply(lambda x: len(str(x).split()))
train_df['question_len'] = train_df['clean_question'].apply(lambda x: len(str(x).split()))

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(train_df['article_len'], bins=30, color='blue', kde=True)
plt.title('Article Word Count')

plt.subplot(1, 2, 2)
sns.histplot(train_df['question_len'], bins=30, color='green', kde=True)
plt.title('Question Word Count')

plt.tight_layout()
plt.show()
```
